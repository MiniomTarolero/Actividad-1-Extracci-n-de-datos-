#Filtro Por Objeto

```python
#Filtro Por objeto 
filtroContinente=df[df['continent']=='South America']
filtroContinente
```

# Filtro Por Filas


```python
#FilasConsecutivas
filtroFilasConsecutivas=df.iloc[100:150,:] 
filtroFilasConsecutivas
```

```python
#Filas no Consecutivas
filtroFilasNoConsecutivas=df.iloc[[1,10,100,122,150,148,27,18],:]
filtroFilasNoConsecutivas
```

```python
#convertir archivo filtrado a csv
filtroFilasConsecutivas.to_csv("filtroFilasConsecutivas.csv")
filtroFilasNoConsecutivas.to_csv("filtroFilasNoConsecutivas.csv")
```

# Filtro Por Columna
